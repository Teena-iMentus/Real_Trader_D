```python
Final_df.drop(columns=["Unnamed: 0"],axis=1,inplace=True)
Final_df['Time Stamp'] = pd.to_datetime(Final_df['Time Stamp'])

# Define the start and end times for filtering
start_time = '2021-10-01 00:00:00'
end_time = '2021-10-01 03:00:00'

# Filter the DataFrame for the specific time range
Final_df = Final_df[(Final_df['Time Stamp'] > start_time) & (Final_df['Time Stamp'] <= end_time)]
```

```python
import pandas as pd
import plotly.graph_objects as go

# Ensure that the 'Time Stamp' column is in datetime format
Final_df['Time Stamp'] = pd.to_datetime(Final_df['Time Stamp'])

# Function to create a candlestick chart with specified annotations
def create_candlestick_chart(df, chart_title, annotations):
    fig = go.Figure(data=[go.Candlestick(
        x=df['Time Stamp'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close'],
        name='Candlestick'
    )])

    # Function to add annotations
    def add_annotations(column, y_col, symbol, color, name):
        signals = df[df[column] != 0]
        if signals.empty:
            # Add default annotation if no signals are present
            fig.add_trace(go.Scatter(x=[df['Time Stamp'].iloc[0]],
                                     y=[df['Low'].min()],
                                     mode='markers',
                                     marker=dict(symbol=symbol, size=10, color=color),
                                     name=name))
        else:
            fig.add_trace(go.Scatter(x=signals['Time Stamp'],
                                     y=signals[y_col],
                                     mode='markers+text',
                                     marker=dict(symbol=symbol, size=10, color=color),
                                     text=signals[column].round(2),  # Display rounded values for hovering
                                     textposition='top center',
                                     name=name))

    # Add specified annotations
    for annotation in annotations:
        add_annotations(*annotation)

    # Update layout
    fig.update_layout(title=chart_title,
                      xaxis_title='Time',
                      yaxis_title='Price',
                      xaxis_rangeslider_visible=True,
                      height=800,
                      width=1500)

    return fig

# First Candlestick Chart
annotations_1 = [
    ('Entry_T', 'Low', 'triangle-up', 'green', 'Entry_T'),
    ('Defence_T', 'High', 'triangle-down', 'red', 'Defence_T'),
    ('Stop_T', 'Low', 'x', 'black', 'Stop_T'),
    ('Target_T', 'High', 'star', 'blue', 'Target_T'),
    ('T_D_Profit', 'High', 'circle', 'green', 'T_D_Profit'),
    ('T_D_Loss', 'Low', 'circle', 'red', 'T_D_Loss'),
]

fig1 = create_candlestick_chart(Final_df, 'Candlestick Chart with TRADER_DATA Annotations', annotations_1)

# Second Candlestick Chart
annotations_2 = [
    ('Entry_C', 'Low', 'triangle-up', 'green', 'Entry_C'),
    ('Defence_C', 'High', 'triangle-down', 'red', 'Defence_C'),
    ('Stop_C', 'Low', 'x', 'black', 'Stop_C'),
    ('Target_C', 'High', 'star', 'blue', 'Target_C'),
    ('C_D_Profit', 'High', 'circle', 'lightgreen', 'C_D_Profit'),
    ('C_D_Loss', 'Low', 'circle', 'lightcoral', 'C_D_Loss'),
]

fig2 = create_candlestick_chart(Final_df, 'Candlestick Chart with CALCULATED_DATA Annotations', annotations_2)

# Third Candlestick Chart
annotations_3 = [
    ('Entry_C', 'Low', 'triangle-up', 'green', 'Entry_C'),
    ('Defence_C', 'High', 'triangle-down', 'red', 'Defence_C'),
    ('Stop_C', 'Low', 'x', 'black', 'Stop_C'),
    ('Target_C', 'High', 'star', 'blue', 'Target_C'),
    ('I_D_Profit', 'High', 'circle', 'blue', 'I_D_Profit'),
    ('I_D_Loss', 'Low', 'circle', 'orange', 'I_D_Loss'),
]

fig3 = create_candlestick_chart(Final_df, 'Candlestick Chart with IDEAL_DATA Annotations', annotations_3)

# Show the plots
fig1.show()
fig2.show()
fig3.show()

# Save the figures as HTML files
fig1.write_html('/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Final_visualization_data/Plotely_Plots/candlestick_chart_TRADER_DATA_annotations_numbers.html')
fig2.write_html('/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Final_visualization_data/Plotely_Plots/candlestick_chart_CALCULATED_DATA_annotations_numbers.html')
fig3.write_html('/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Final_visualization_data/Plotely_Plots/candlestick_chart_IDEAL_DATA_annotations_numbers.html')
```

```python
import pandas as pd
import plotly.graph_objects as go

# Ensure that the 'Time Stamp' column is in datetime format
Final_df['Time Stamp'] = pd.to_datetime(Final_df['Time Stamp'])

# Function to create a candlestick chart with specified annotations
def create_candlestick_chart(df, chart_title, entry_column, hover_columns, symbol, color, name):
    fig = go.Figure(data=[go.Candlestick(
        x=df['Time Stamp'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close'],
        name='Candlestick'
    )])

    # Add entry annotations with hover text
    signals = df[df[entry_column] != 0]
    if not signals.empty:
        hover_text = signals.apply(lambda row: '<br>'.join(
            f'{col}: {row[col]}' for col in hover_columns if row[col] != 0), axis=1)

        fig.add_trace(go.Scatter(x=signals['Time Stamp'],
                                 y=signals['Low'],  # Assuming entry is related to the low price
                                 mode='markers',
                                 marker=dict(symbol=symbol, size=10, color=color),
                                 text=hover_text,  # Custom hover text
                                 hoverinfo='text',
                                 name=name))

    # Update layout
    fig.update_layout(title=chart_title,
                      xaxis_title='Time',
                      yaxis_title='Price',
                      xaxis_rangeslider_visible=True,
                      height=800,
                      width=1500)

    return fig

# Define columns to include in hover text
hover_columns_1 = ['Entry_T', 'Defence_T', 'Stop_T', 'Target_T', 'T_D_Profit', 'T_D_Loss']

# First Candlestick Chart with Entry_T
fig1 = create_candlestick_chart(
    Final_df,
    'Candlestick Chart with TRADER_DATA Annotations',
    entry_column='Entry_T',
    hover_columns=hover_columns_1,
    symbol='triangle-up',
    color='green',
    name='Entry_T'
)

# Define columns to include in hover text for the second chart
hover_columns_2 = ['Entry_C', 'Defence_C', 'Stop_C', 'Target_C', 'C_D_Profit', 'C_D_Loss']

# Second Candlestick Chart with Entry_C
fig2 = create_candlestick_chart(
    Final_df,
    'Candlestick Chart with CALCULATED_DATA Annotations',
    entry_column='Entry_C',
    hover_columns=hover_columns_2,
    symbol='triangle-up',
    color='green',
    name='Entry_C'
)

# Define columns to include in hover text for the third chart
hover_columns_3 = ['Entry_C', 'Defence_C', 'Stop_C', 'Target_C', 'I_D_Profit', 'I_D_Loss']

# Third Candlestick Chart with Entry_C for IDEAL_DATA
fig3 = create_candlestick_chart(
    Final_df,
    'Candlestick Chart with IDEAL_DATA Annotations',
    entry_column='Entry_C',
    hover_columns=hover_columns_3,
    symbol='triangle-up',
    color='green',
    name='Entry_C'
)

# Show the plots
fig1.show()
fig2.show()
fig3.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Ensure that the 'Time Stamp' column is in datetime format
Final_df['Time Stamp'] = pd.to_datetime(Final_df['Time Stamp'])

# Function to create a candlestick chart with specified annotations
def create_candlestick_chart(df, chart_title, annotations):
    fig = go.Figure(data=[go.Candlestick(
        x=df['Time Stamp'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close'],
        name='Candlestick'
    )])

    # Function to add annotations
    def add_annotations(column, y_col, symbol, color, name, hover_text):
        signals = df[df[column] != 0]
        if not signals.empty:
            fig.add_trace(go.Scatter(
                x=signals['Time Stamp'],
                y=signals[y_col],
                mode='markers',
                marker=dict(symbol=symbol, size=10, color=color),
                hovertext=[hover_text.format(val) for val in signals[y_col]],
                hoverinfo='text',
                name=name
            ))

    # Define specific hover text for each annotation
    hover_texts = {
        'Entry_T': 'Entry_T Value: {:.2f}',
        'Defence_T': 'Defence_T Value: {:.2f}',
        'Stop_T': 'Stop_T Value: {:.2f}',
        'Target_T': 'Target_T Value: {:.2f}',
        'T_D_Profit': 'T_D_Profit Value: {:.2f}',
        'T_D_Loss': 'T_D_Loss Value: {:.2f}',
        'Entry_C': 'Entry_C Value: {:.2f}',
        'Defence_C': 'Defence_C Value: {:.2f}',
        'Stop_C': 'Stop_C Value: {:.2f}',
        'Target_C': 'Target_C Value: {:.2f}',
        'C_D_Profit': 'C_D_Profit Value: {:.2f}',
        'C_D_Loss': 'C_D_Loss Value: {:.2f}',
        'I_D_Profit': 'I_D_Profit Value: {:.2f}',
        'I_D_Loss': 'I_D_Loss Value: {:.2f}'
    }

    # Add specified annotations
    for column, y_col, symbol, color, name in annotations:
        hover_text = hover_texts.get(name, '{:.2f}')
        add_annotations(column, y_col, symbol, color, name, hover_text)

    # Update layout
    fig.update_layout(
        title=chart_title,
        xaxis_title='Time',
        yaxis_title='Price',
        xaxis_rangeslider_visible=True,
        height=800,
        width=1500
    )

    return fig

# First Candlestick Chart
annotations_1 = [
    ('Entry_T', 'Low', 'triangle-up', 'green', 'Entry_T'),
    ('Defence_T', 'High', 'triangle-down', 'red', 'Defence_T'),
    ('Stop_T', 'Low', 'x', 'black', 'Stop_T'),
    ('Target_T', 'High', 'star', 'blue', 'Target_T'),
    ('T_D_Profit', 'High', 'circle', 'green', 'T_D_Profit'),
    ('T_D_Loss', 'Low', 'circle', 'red', 'T_D_Loss'),
]

fig1 = create_candlestick_chart(Final_df, 'Candlestick Chart with TRADER_DATA Annotations', annotations_1)

# Second Candlestick Chart
annotations_2 = [
    ('Entry_C', 'Low', 'triangle-up', 'green', 'Entry_C'),
    ('Defence_C', 'High', 'triangle-down', 'red', 'Defence_C'),
    ('Stop_C', 'Low', 'x', 'black', 'Stop_C'),
    ('Target_C', 'High', 'star', 'blue', 'Target_C'),
    ('C_D_Profit', 'High', 'circle', 'lightgreen', 'C_D_Profit'),
    ('C_D_Loss', 'Low', 'circle', 'lightcoral', 'C_D_Loss'),
]

fig2 = create_candlestick_chart(Final_df, 'Candlestick Chart with CALCULATED_DATA Annotations', annotations_2)

# Third Candlestick Chart
annotations_3 = [
    ('Entry_C', 'Low', 'triangle-up', 'green', 'Entry_C'),
    ('Defence_C', 'High', 'triangle-down', 'red', 'Defence_C'),
    ('Stop_C', 'Low', 'x', 'black', 'Stop_C'),
    ('Target_C', 'High', 'star', 'blue', 'Target_C'),
    ('I_D_Profit', 'High', 'circle', 'blue', 'I_D_Profit'),
    ('I_D_Loss', 'Low', 'circle', 'orange', 'I_D_Loss'),
]

fig3 = create_candlestick_chart(Final_df, 'Candlestick Chart with IDEAL_DATA Annotations', annotations_3)

# Show the plots
fig1.show()
fig2.show()
fig3.show()

# Save the figures as HTML files
fig1.write_html('/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Final_visualization_data/Plotely_Plots/candlestick_chart_TRADER_DATA_annotations.html')
fig2.write_html('/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Final_visualization_data/Plotely_Plots/candlestick_chart_CALCULATED_DATA_annotations.html')
fig3.write_html('/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Final_visualization_data/Plotely_Plots/candlestick_chart_IDEAL_DATA_annotations.html')
```
