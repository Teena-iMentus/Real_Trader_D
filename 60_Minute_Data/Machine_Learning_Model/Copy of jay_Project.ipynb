```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42)
}
# Initialize dictionary to store accuracy scores
results_accuracy = {}

# Train and evaluate each model, storing only accuracy scores
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on the test data
    y_pred = model.predict(X_test)
    
    # Compute accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store accuracy score in dictionary
    results_accuracy[model_name] = accuracy

# Convert results to DataFrame
df_accuracy = pd.DataFrame(results_accuracy.items(), columns=['Model', 'Accuracy'])

# Display the DataFrame
print("Accuracy Scores:")
print(df_accuracy)
```

```python
# Drop unnecessary columns
data.drop(columns=['Unnamed: 0'])
```

```python
# Reorder the columns
columns_order = ['Date', 'Open', 'High', 'Low', 'Close', 'BearControl4', 'BearControl3', 'BearControl2', 'BearControl1', 'BearControl0', 'BullControl0', 'BullControl1', 'BullControl2', 'BullControl3', 'BullControl4']
data = data[columns_order]
```

```python
from sklearn.preprocessing import MinMaxScaler
# Drop the first column ('Date')
data = data.iloc[:, 1:]  # Assuming the first column is 'Date'
data.dtypes
# Extract numerical columns for normalization
numerical_columns = data.columns

# Initialize the scaler
scaler = MinMaxScaler()

# Normalize the numerical columns
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python
data.shape
```

```python
type(data)
```

```python
# Function to create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :])  # Use all columns including 'Date'
        y.append(data[i+seq_length, :])    # Predicting all columns including 'Date'
    return np.array(X), np.array(y)

# Convert DataFrame to numpy array
data = data.values

# Create sequences
seq_length = 10  # Example sequence length
X, y = create_sequences(data, seq_length)

# Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from keras.layers import GRU, Dense
from tensorflow.keras.optimizers import Adam

```

```python
# Define the LSTM model
model = Sequential()
model.add(LSTM(units=50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(units=y_train.shape[1]))  # Output layer predicts all columns including 'Date'

# Compile the model
model.compile(optimizer=Adam(), loss='mse')
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# # Fit the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on test data
loss = model.evaluate(X_test, y_test)

print(f'Test Loss: {loss}')
```

```python
# Make predictions on training and test data
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculate MSE and MAE for training data
train_mse = mean_squared_error(y_train, train_predictions)
train_mae = mean_absolute_error(y_train, train_predictions)

# Calculate MSE and MAE for test data
test_mse = mean_squared_error(y_test, test_predictions)
test_mae = mean_absolute_error(y_test, test_predictions)

print(f'Training MSE: {train_mse}, MAE: {train_mae}')
print(f'Test MSE: {test_mse}, MAE: {test_mae}')
```
