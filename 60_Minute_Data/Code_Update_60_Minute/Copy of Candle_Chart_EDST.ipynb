```python
#importent libraries
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
pd.set_option('display.max_columns',40)
```

```python
ES_5_Min = pd.read_csv("/content/drive/MyDrive/Real_Trader_EDA_R1/Data/New_5min_data (2).csv")
ES_5_Min.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Convert 'Time Stamp' column to datetime
ES_5_Min['Time Stamp'] = pd.to_datetime(ES_5_Min['Time Stamp'])

# Filter data for the specific date and time range
start_datetime = '2019-06-24 13:15:00'
end_datetime = '2019-06-24 15:40:00'
mask = (ES_5_Min['Time Stamp'] >= start_datetime) & (ES_5_Min['Time Stamp'] <= end_datetime)
filtered_df = ES_5_Min.loc[mask]
```

```python
filtered_df.shape
```

```python
#set_index time_stamp column
filtered_df.set_index('Time Stamp',inplace=True)
```

```python
print(filtered_df.columns)
print(filtered_df.shape)
```

```python
filtered_df
```

```python
# Create a subplot with two rows
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('BuyersInControlPrice', 'SellersInControlPrice'),
    shared_xaxes=True
)

# Add the candlestick chart to both subplots
fig.add_trace(go.Candlestick(
    x=filtered_df.index,
    open=filtered_df['Open'],
    high=filtered_df['High'],
    low=filtered_df['Low'],
    close=filtered_df['Close'],
    name='Candlestick'
), row=1, col=1)

fig.add_trace(go.Candlestick(
    x=filtered_df.index,
    open=filtered_df['Open'],
    high=filtered_df['High'],
    low=filtered_df['Low'],
    close=filtered_df['Close'],
    name='Candlestick'
), row=2, col=1)

# Add horizontal lines and annotations for BuyersInControlPrice
for index, row in filtered_df.iterrows():
    if row['BuyersInControlPrice'] != 0:
        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['BuyersInControlPrice'], row['BuyersInControlPrice']],
            mode='lines+text',
            line=dict(color='green', width=2),
            text=[f'Entry: {row["BuyersInControlPrice"]}', f'Entry: {row["BuyersInControlPrice"]}'],
            textposition='bottom left',
            hovertext=f'Entry: {row["BuyersInControlPrice"]}',
            hoverinfo='text',
            name='Entry',
            showlegend=False
        ), row=1, col=1)

        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['Defence'], row['Defence']],
            mode='lines+text',
            line=dict(color='blue', width=2),
            text=[f'Defence: {row["Defence"]}', f'Defence: {row["Defence"]}'],
            textposition='bottom left',
            hovertext=f'Defence: {row["Defence"]}',
            hoverinfo='text',
            name='Defence',
            showlegend=False
        ), row=1, col=1)

        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['Stop'], row['Stop']],
            mode='lines+text',
            line=dict(color='purple', width=2),
            text=[f'Stop: {row["Stop"]}', f'Stop: {row["Stop"]}'],
            textposition='bottom left',
            hovertext=f'Stop: {row["Stop"]}',
            hoverinfo='text',
            name='Stop',
            showlegend=False
        ), row=1, col=1)

        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['Target'], row['Target']],
            mode='lines+text',
            line=dict(color='red', width=2),
            text=[f'Target: {row["Target"]}', f'Target: {row["Target"]}'],
            textposition='bottom left',
            hovertext=f'Target: {row["Target"]}',
            hoverinfo='text',
            name='Target',
            showlegend=False
        ), row=1, col=1)

# Add horizontal lines and annotations for SellersInControlPrice
for index, row in filtered_df.iterrows():
    if row['SellersInControlPrice'] != 0:
        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['SellersInControlPrice'], row['SellersInControlPrice']],
            mode='lines+text',
            line=dict(color='orange', width=2),
            text=[f'Entry: {row["SellersInControlPrice"]}', f'Entry: {row["SellersInControlPrice"]}'],
            textposition='bottom left',
            hovertext=f'Entry: {row["SellersInControlPrice"]}',
            hoverinfo='text',
            name='Entry',
            showlegend=False
        ), row=2, col=1)

        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['Defence'], row['Defence']],
            mode='lines+text',
            line=dict(color='cyan', width=2),
            text=[f'Defence: {row["Defence"]}', f'Defence: {row["Defence"]}'],
            textposition='bottom left',
            hovertext=f'Defence: {row["Defence"]}',
            hoverinfo='text',
            name='Defence',
            showlegend=False
        ), row=2, col=1)

        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['Stop'], row['Stop']],
            mode='lines+text',
            line=dict(color='magenta', width=2),
            text=[f'Stop: {row["Stop"]}', f'Stop: {row["Stop"]}'],
            textposition='bottom left',
            hovertext=f'Stop: {row["Stop"]}',
            hoverinfo='text',
            name='Stop',
            showlegend=False
        ), row=2, col=1)

        fig.add_trace(go.Scatter(
            x=[filtered_df.index.min(), filtered_df.index.max()],
            y=[row['Target'], row['Target']],
            mode='lines+text',
            line=dict(color='red', width=2),
            text=[f'Target: {row["Target"]}', f'Target: {row["Target"]}'],
            textposition='bottom left',
            hovertext=f'Target: {row["Target"]}',
            hoverinfo='text',
            name='Target',
            showlegend=False
        ), row=2, col=1)

# Update layout
fig.update_layout(
    title='Candlestick Chart with BuyersInControlPrice and SellersInControlPrice',
    xaxis_title='Time',
    yaxis_title='Price',
    xaxis_rangeslider_visible=False,
    xaxis_tickformat='%Y-%m-%d %H:%M:%S',  # Format x-axis to show date and time
    xaxis_tickangle=-45,  # Rotate x-axis labels for better readability
    height=800  # Increase height to accommodate two subplots
)

# Save the plot as an image file
#fig.write_image('/content/drive/MyDrive/Real_Trader_EDA_R1/Data/Candlestick_Chart_Subplots_Hover.png')

# Show the plot (for interactive use)
fig.show()
```
