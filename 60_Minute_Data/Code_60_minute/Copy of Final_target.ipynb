```python
df_ideal = pd.read_csv("/content/drive/MyDrive/Real_Trader_EDA_R1/Target_Optimizaton_data/Copy of ideal_target.csv")
df_calculated = pd.read_csv("/content/drive/MyDrive/Real_Trader_EDA_R1/Target_Optimizaton_data/Copy of New_5min_data (2).csv")
df_trader = pd.read_csv("/content/drive/MyDrive/Real_Trader_EDA_R1/Target_Optimizaton_data/Copy of ES 09-24_Minute_5.csv")
```

```python
df_trader['Time Stamp'] = pd.to_datetime(df_trader['Time Stamp'])
df_ideal['Time Stamp'] = pd.to_datetime(df_ideal['Time Stamp'])
df_calculated['Time Stamp'] = pd.to_datetime(df_calculated['Time Stamp'])
```
